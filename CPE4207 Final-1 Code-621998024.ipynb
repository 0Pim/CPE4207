```python
# จงเขียนโปรแกรมหรืออัลกอริทึมสำหรับจัดตารางงานให้แก่ทีมช่างเทคนิค (แบบรายสัปดําห์) กำหนดให้
# องค์ประกอบของการจัดตารางงานดังนี้
# กำหนดให้ทีมช่างเทคนิคมี 3 ทีม โดยช่างทำงาน 5 วันต่อสัปดาห์ และแต่ละวันทำงานแบ่งเป็นสองช่วง
# คือเช้าและบ่าย โดยนักศึกษาเขียนโปรแกรมจัดตารางงานให้แก่ ทีมช่างทุกทีมอย่างเท่าเทียมกัน ด้วยวิธี
# การแบบ Genetic Algorithm โดยให้นักศึกษากำหนดปริมาณงานของบริษัทแต่ละสัปดาห์ไม่น้อยกว่า 
# 20 ภาระงาน

import numpy as np
from numpy import random
```

```python
class Teams:
    def __init__(self):
        self.Tname = np.array([
            ["A"],
            ["B"],
            ["C"]
        ])

        self.SlotTeam = np.array([
            ["xxx"], ["xxx"],
            ["xxx"], ["xxx"],
            ["xxx"], ["xxx"],
            ["xxx"], ["xxx"],
            ["xxx"], ["xxx"]
        ])
        self.SlotTeamWeek = np.array([self.SlotTeam for _ in range(len(self.Tname))]
                                     )
    def getTeamsName(self):
        return self.Tname

    def getTeamWeek(self):
        return self.SlotTeamWeek
```

```python
class JobsWorkload:
    def __init__(self):
        self.CWL = np.array([ # Course, Team, TimeSlot
                                ["001", -1, -1],
                                ["002", -1, -1],
                                ["003", -1, -1],
                                ["004", -1, -1],
                                ["005", -1, -1],
                                ["006", -1, -1],
                                ["007", -1, -1],
                                ["008", -1, -1],
                                ["009", -1, -1],
                                ["010", -1, -1],
                                ["011", -1, -1],
                                ["012", -1, -1],
                                ["013", -1, -1],
                                ["014", -1, -1],
                                ["015", -1, -1],
                                ["016", -1, -1],
                                ["017", -1, -1],
                                ["018", -1, -1],
                                ["019", -1, -1],
                                ["020", -1, -1]
        ])

    def getJobsWorkload(self):
        return self.CWL
```

```python
def Scheduling(Cwl, Tw):
    for i in range(len(Cwl)):
        rdTeam = random.randint(0, len(Tw))
        rdSlot = random.randint(0, len(Tw[0])-1)
        while (Tw[rdTeam, rdSlot] != "xxx"):
            rdTeam = random.randint(0, len(Tw)-1)
            rdSlot = random.randint(0, len(Tw[0])-1)
        Cwl[i, 1] = rdTeam
        Cwl[i, 2] = rdSlot
        Tw[int(Cwl[i, 1]), int(Cwl[i, 2])] = Cwl[i, 0]
```

```python
def ComputeFittnessValue(Cwl):
    value = 50
    num = [Cwl[i][1] for i in range(0, len(Cwl))]
    u, counts = np.unique(num, return_counts=True)
    sum = (counts[0]+counts[1]+counts[2])/3
    Max = int(np.ceil(sum))
    Min = int(np.floor(sum))

    if counts[0] == Max or counts[0] == Min:
        if counts[1] == Max or counts[1] == Min:
            if counts[2] == Max or counts[2] == Min:
                value = 50
            else:
                value = 0
        else:
            value = 0
    else:
        value = 0

    return value, counts
```

```python
def Display(T, F, C):
    Tw = T.getTeamWeek()
    Tm = T.Tname
    print("ตารางช่างทำงาน 5 วันต่อสัปดาห์ ")
    print('')
    for i in range(len(Tw)):
        print("Team", Tm[i][0], "มี", C[i], "งาน")
        for sl in range(len(Tw[i])):
            print(Tw[i, sl], end=" ")
            if sl % 2 == 1:
                print()
        print()
    print(F)

```

```python
def DisplayJobsWorkLoad(Cwl):
    print(Cwl)
```

```python
# np.random.seed(2)
fitness = 0
while (fitness != 50):
    T = Teams()
    J = JobsWorkload()
    Scheduling(J.getJobsWorkload(), T.getTeamWeek())
    fitness, counts = ComputeFittnessValue(J.getJobsWorkload())

    if (fitness == 50):
        print("ปริมาณงานบริษัทมีทั้งหมด :", len(J.getJobsWorkload()))
        print('')
        
        Display(T, fitness, counts)
        DisplayJobsWorkLoad(J.getJobsWorkload())
```
